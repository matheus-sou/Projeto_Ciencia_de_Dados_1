## **02.LIMPANDO OS DADOS AUSENTES.**

### **02.a.Subistituindo os dados ausentes.**

Calcule a **moda** e use o método **`fillna()`** para substituir os dados ausentes na coluna **FORMA_DE_PAGAMENTO**.

```python
# Substituindo os valores ausentes

# 01.Calculando a moda para a coluna FORMA_DE_PAGAMENTO e armazenando em moda_FORMAS_DE_PAGAMENTO
moda_FORMAS_DE_PAGAMENTO = df.FORMA_DE_PAGAMENTO.mode()[0]

# 02.Preenchendo os dados ausentes com o valor da moda_FORMAS_DE_PAGAMENTO.
df.FORMA_DE_PAGAMENTO.fillna(moda_FORMAS_DE_PAGAMENTO, inplace=True)

# 03.Verificando a alteração.
df.isnull().sum()
```

#### **(EXTRA) 03.f.Qual região geográfica há mais venda?.**

Como ainda iremos aprender a fazer gráficos geográficos com a biblioteca **Folium** apenas execute a célula de código a seguir para criar o mapa.

```python
# 01.Importar o Folium para o Jupyter Notebook
import folium
from folium import plugins

# 02.Extraindo valores do DataFrame para variáveis LATITUDE e LONGITUDE, e armazendo em uma lista
coordenadas=[]
for lat,lng in zip(df.LATITUDE.values, df.LONGITUDE.values):
  coordenadas.append([lat,lng])

# 03.Criando um mapa
mapa = folium.Map(location=[-23.519016 ,-47.439082], zoom_start=12)

# 04.Adicionar o mapa de temperatura
mapa.add_child(plugins.HeatMap(coordenadas))
```
